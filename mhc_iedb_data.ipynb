```python
import pandas as pd

df = pd.read_csv('human_mhc_classI.csv', low_memory=False)
HLA_list = df['mhc_allele'].unique().tolist()

with open('mhc_classI_hla_ids.txt', 'w') as f:
    for hla in HLA_list:
        hla = hla.replace('HLA-','')
        f.write(hla + '\n')
```

```python
from Bio import SeqIO
import json

hla_ids = open('mhc_classI_hla_ids.txt','r')
hla_id = hla_ids.read().split()
hla_ids.close()

hla_dict = {}
for hla in hla_id:
    for seq_record in SeqIO.parse("hla_prot_trim.fasta", "fasta"):
        if hla in seq_record.id:
            hla_dict['HLA-'+hla] = str(seq_record.seq)
            break



json.dump(hla_dict, open( "hla_allele_classI.json", 'w' ) )
```
